## 9. Handle Imbalanced Dataset 

```python
# 1. Encode the categorical features.
# 2. Add the encoded categorical features to the dataset.
# 3. Apply smote transformation to the dataset.
# 4. Store the dataset as clean_dataset
```

```python
df_copy = df_clean
```

```python
for cat in cat_features:
    print(f'{cat}: {df_clean[cat].unique()}')
```

```python
HeartDisease =      ['No' , 'Yes'] 
Smoking =           ['No',  'Yes']
AlcoholDrinking =   ['No' , 'Yes']
Stroke =            ['No' , 'Yes']
DiffWalking =       ['No' 'Yes']
Sex =               ['Female' , 'Male']
AgeCategory =       ['18-24',  '25-29',  '30-34',  '35-39',  '40-44' , '45-49' , '50-54',  '55-59', '60-64',  '65-69' , '70-74',  '75-79' , '80 or older']
Race =              ['White' , 'Black' , 'Asian' , 'American Indian/Alaskan Native',  'Hispanic','Other']
Diabetic =          ['No' , 'No, borderline diabetes' , 'Yes (during pregnancy)' , 'Yes']
PhysicalActivity =  ['No',  'Yes']
GenHealth =         ['Poor',  'Fair' , 'Good',  'Very good',  'Excellent']
Asthma =            ['No',  'Yes']
KidneyDisease =     ['No' , 'Yes']
SkinCancer =        ['No' , 'Yes']
```

```python
df_cat = df_clean[cat_features]
df_cat.head(10)
```

```python
from sklearn.preprocessing import  OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

df_cat_encoded = ordinal_encoder.fit_transform(df_cat)
```

```python
df_cat_encoded_ = pd.DataFrame(df_cat_encoded)
df_cat_encoded_.columns = cat_features
df_cat_encoded_.head(10)
```

```python
df_cat_encoded_.shape
```

```python
for cat in cat_features:
    df_clean[cat] = df_cat_encoded_[cat]

df_clean.head(10)
```

```python
# Now we will apply smotenc technique
```

```python
X = df_clean.drop(['HeartDisease'],axis=1)
y = df_clean['HeartDisease']
```

```python
cat_indices = []
for i in range(len(list(X.columns))):
    if list(X.columns)[i] in cat_features:
        cat_indices.append(i)

cat_indices
```

```python
X.columns
```

```python
y
```

```python
# Apply SMOTE-NC
from imblearn.over_sampling import SMOTENC

smote_nc = SMOTENC(categorical_features=cat_indices)
X_resampled, y_resampled = smote_nc.fit_resample(X, y)

```

```python
X_resampled.head(10)
```

```python
X_resampled.tail(10)
```

```python
for cat in cat_features:
    if cat != 'HeartDisease':
        print(f'{cat}: {X_resampled[cat].unique()}\n')
```
